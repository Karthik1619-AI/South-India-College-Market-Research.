```python
import matplotlib.pyplot as plt

avg_placement = (
    df.groupby("State")["Placement Statistics"]
    .mean()
    * 100
).sort_values(ascending=False)

plt.figure(figsize=(8,5))

ax = avg_placement.plot(kind="bar")

plt.title("Average Placement Percentage by State")
plt.xlabel("State")
plt.ylabel("Placement (%)")

plt.ylim(70, 80)

for i, v in enumerate(avg_placement):
    ax.text(i, v + 0.1, f"{v:.2f}%", ha="center")

plt.tight_layout()
plt.show()
```
